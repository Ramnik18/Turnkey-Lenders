### Dataset Description - 
The columns/features in the given dataset are as follows: 
1. ID: ID of each client
2. AMT: Amount of given credit in dollars (includes individual and family/supplementary credit)
3. GENDER: Gender (1=male, 2=female)
4. EDUCATION: (1=graduate school, 2=university, 3=high school, 4=others, 5=unknown)
5. MARITAL STATUS: Marital status (1=married, 2=single, 3=others)
6. AGE: Age in years
7. REPAY_SEP: Repayment status in September, 2005 (0=pay duly, 1=payment delay for one month, 2=payment delay for two months, â€¦ 8=payment delay for eight months, 9=payment delay for nine months and above)
8. REPAY_AUG: Repayment status in August, 2005 (scale same as above)
9. REPAY_JUL: Repayment status in July, 2005 (scale same as above)
10. REPAY_JUN: Repayment status in June, 2005 (scale same as above)
11. REPAY_MAY: Repayment status in May, 2005 (scale same as above)
12. REPAY_APR: Repayment status in April, 2005 (scale same as above)
13. AMTBILL_SEP: Amount of bill statement in September, 2005
14. AMTBILL_AUG: Amount of bill statement in August, 2005
15. AMTBILL_JUL: Amount of bill statement in July, 2005
16. AMTBILL_JUN: Amount of bill statement in June, 2005
17. AMTBILL_MAY: Amount of bill statement in May, 2005
18. AMTBILL_APR: Amount of bill statement in April, 2005
19. PRE_SEP: Amount of previous payment in September, 2005
20. PRE_AUG: Amount of previous payment in August, 2005
21. PRE_JUL: Amount of previous payment in July, 2005
22. PRE_JUN: Amount of previous payment in June, 2005
23. PRE_MAY: Amount of previous payment in May, 2005
24. PRE_APR: Amount of previous payment in April, 2005
25. DEF_AMT: Default payment (1=yes, 0=no)
